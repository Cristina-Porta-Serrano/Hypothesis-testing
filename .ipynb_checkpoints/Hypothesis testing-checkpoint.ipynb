Volem veure si la mitjana dels guanys de tots els països és 1000 milions d'euros o més, agafant una alpha de 0'05. Fem servir el one sample t-test i entre els parèntesis posem el dataset i la mitjana esperada a la hipòtesis nul·la).

<u>*Exercice 2*<u>

*Continua amb el conjunt de dades de tema esportiu que t'agradi i selecciona dos altres atributs del conjunt de dades. Calcula els p-valors i digues si rebutgen la hipòtesi nul·la agafant un alfa de 5%.*

Per realitzar aquest anàlisis fem servir el Chi-Square test, que és un test estadístic no-paramètric que ens ajuda a entendre les relacions entre les variables categòriques del dataset. Fent-lo servir, podem estimar el nivell de correlació entre elles, i això ens ajuda a analitzar la dependència d'una categoria de la variable respecte una altra categoria independent de la variable. 

La hipòtesi nula la podem definir així: les variables agrupades no tenen correlació entre elles. 
La hipòtesi alternativa quedaria de la següent manera: les variables estan associades entre elles i tenen correlació. 

Una Contingency table (o crosstab) s'usa en estadística per resumir la relació entre 2 o més variables categòriques. En aquest cas, volem saber si hi ha relació entre l'esport que practiquen els esportistes més rics i la seva nacionalitat.

```python
richest_athletes['Sport'].unique()
```

```python
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='boxing', value='Boxing')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='basketball', value='Basketball')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='golf', value='Golf')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='baseball', value='Baseball')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='F1 Motorsports', value='F1 racing')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='F1 racing', value='Auto racing')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='tennis', value='Tennis')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='Auto Racing (Nascar)', value='NASCAR')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='soccer', value='Soccer')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='Auto racing', value='Auto Racing')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='ice hockey', value='Ice Hockey')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='NBA', value='Basketball')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='NFL', value='American Football / Baseball')
richest_athletes['Sport'] = richest_athletes['Sport'].replace(to_replace='American Football / Baseball', value='American Football')
```

```python
richest_athletes
```

```python
richest_athletes['Sport'].unique()
```

```python
contingency_table=pd.crosstab(richest_athletes["Sport"],richest_athletes["Nationality"])
print('CONTINGENCY TABLE:\n',contingency_table)
```

Fem una taula amb els valors esperats, que es calcula de la següent manera:

Un valor esperat és la freqüència teòricament esperada d'una cel·la, suposant que els grups d'estudi són independents. Per calcular un valor esperat per a una cel·la d'una taula de contingències, calculem el producte del total de fila i el total de columna per a la fila i la columna que contenen aquesta cel·la i, a continuació, dividim aquest producte calculat pel total general.

```python
#Observed Values
Observed_Values = contingency_table.values 
#print("Observed Values :-\n",Observed_Values)
b=stats.chi2_contingency(contingency_table)
Expected_Values = b[3]
print("Expected Values:\n",Expected_Values)
```

Fem una taula amb els valors de chi-square, que es calcula de la següent manera:

Agafem el valor observat (o) - el valor esperat (e) i l'elevem al quadrat. Dividim el resultat entre el resultat esperat.

Calculem també el dof (degree of freedom que es calcula (nº de files – 1) * (nº de columnes – 1)), el p-value i el critical value (que es pot extreure d'una taula o fer servir la funció chi2.ppf). El significance level és la alpha, agafem 0'05.

```python
no_of_rows=len(contingency_table)
no_of_columns=len(contingency_table)
dof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:",dof)
alpha = 0.05

chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
print("chi-square statistic:",chi_square_statistic)
critical_value=chi2.ppf(q=1-alpha,df=dof)
print('critical_value:',critical_value)
p_value=1-chi2.cdf(x=chi_square_statistic,df=dof)
print('p-value:',p_value)
print('Significance level: ',alpha)

if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

Com que el p-value ens dona 0'99 i és més gran que la alpha de 0'05, agafem la hipòtesi nul·la perquè NO que hi ha relació entre les variables.
